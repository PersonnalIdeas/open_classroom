```python
import pandas as pd
import missingno as m
```

```python
df=pd.read_csv('fr.openfoodfacts.org.products.csv',sep='\t',low_memory=False)
```

```python
m.matrix(df)
```

```python
#taux de remplissage d'une colonne
print('alcohol_100g : '+str((len(df)-sum(df.alcohol_100g.isnull()))/len(df)))
```

```python
#remplacer des valeurs manquantes d'une colonne :
df.alcohol_100g.fillna(0,inplace=True)
#remplacer les valeurs abherantes d'une colonne. ( on remplace tous les 0 en 10000)
df.ix[df.alcohol_100g == 0, 'alcohol_100g'] = 10000
#supprimer une colonne
del df['alcohol_100g']
```
