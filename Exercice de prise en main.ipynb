## Ici : Je regarde le taux de remplissage des colonnes, si celui-ci est inferieur a 35%. Je supprime la colonne. Sinon je remplis les valeurs manquantes.

```python
# Detection des valeurs manquantes
```

```python
len(df.Cabin.dropna())/len(df)
```

<b> La variable Cabin n'est remplie qu'a hauteur de 22 %. On la supprime.


```python
del df['Cabin']
```

```python
print('Il y a ' + str( int(100-100*len( df.Age.dropna() ) / len(df))) + '% de valeurs manquantes dans la colonne Age')
```

```python
plt.boxplot(df.Age.dropna()) #plot sans les valeurs manquantes
plt.title('Repartition des valeurs de l\'Age')
plt.ylabel('Age')
plt.xticks([])
plt.show()
```

Ici 30 semble une bonne valeur.

```python
df.Age.fillna(30,inplace=True)
```

```python
plt.boxplot(df.Age)
plt.title('Repartition des valeurs de l\'Age avec les valeurs manquantes.')
plt.ylabel('Age')
plt.xticks([])
plt.show()
```
